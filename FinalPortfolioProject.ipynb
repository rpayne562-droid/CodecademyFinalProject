```python
sns.barplot(data=AutoInsuranceData, x='Location Code', y='Total Claim Amount', errorbar=None)
plt.title('Which location has the lowest claim amount?')
plt.savefig('figure1.png')
plt.show()
```

**Rural has the lowest claim amounts, followed by urban.**

```python
sns.barplot(data=AutoInsuranceData, x='Vehicle Class', y='Total Claim Amount', errorbar=None)
plt.title('Which vehicle class has the lowest claim amount?')
plt.xticks(rotation=20)
plt.savefig('figure2.png')
plt.show()
```

**Non-luxury cars have much lowest claim amounts than luxury vehicles do.**

```python
sns.barplot(data=AutoInsuranceData, x='Vehicle Size', y='Total Claim Amount', errorbar=None)
plt.title('Which vehicle size has the lowest claim amounts?')
plt.savefig('figure3.png')
plt.show()
```

**Small vehicles have the highest claim amounts with medium and large vehicles approximately the same.**

```python
sns.barplot(data=AutoInsuranceData, x='Education', y='Total Claim Amount', errorbar=None)
plt.title('Which education level has the lowest claim amount?')
plt.xticks(rotation=10)
plt.savefig('figure4.png')
plt.show()
```

**Master and Doctor have the lowest claim amount while high school or below is the highest.  The more education they have, the lower the claim amount.**

```python
sns.barplot(data=AutoInsuranceData, x='Marital Status', y='Total Claim Amount', errorbar=None)
plt.title('Which marital status has the lowest claim amount?')
plt.savefig('figure5.png')
plt.show()
```

**Single people have the highest claim amounts with married and divorced people approximately the same.**

```python
sns.barplot(data=AutoInsuranceData, x='EmploymentStatus', y='Total Claim Amount', errorbar=None)
plt.title('Which employment status has the lowest claim amount?')
plt.savefig('figure6.png')
plt.show()
```

**Employed policyholders have the lowest claim amount.  Medical leave and disabled are about the same.**

```python
sns.scatterplot(data=AutoInsuranceData.query('Income > 0'), x='Income', y='Total Claim Amount')
plt.title('Is there any relationship between income and total claim amount?')
plt.savefig('figure7.png')
plt.show()
```

**Income doesn't matter.  Whether it is 10,000 or 100,000, all of them have the same claim amounts.**

# Conclusion
The best risks are:
- Rural policyholders
- Non-luxury vehicles especially sedans
- Medium and large size vehicles
- Policyholders with a master or higher education level
- Married or divorced policyholders
- Policyholders who are employed
